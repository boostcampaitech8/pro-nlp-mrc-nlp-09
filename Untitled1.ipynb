```python

from datasets import load_from_disk

ds = load_from_disk("./data/train_dataset/negative")

print("Total:", len(ds))

missing = []
not_list = []
non_str = []
length_zero = []

for i in range(len(ds)):
    row = ds[i]
    if "negatives" not in row:
        missing.append(i)
    elif not isinstance(row["negatives"], list):
        not_list.append(i)
    elif any(not isinstance(x, str) for x in row["negatives"]):
        non_str.append(i)
    elif len(row["negatives"]) == 0:
        length_zero.append(i)

print("Missing negatives:", missing[:10], "count:", len(missing))
print("Non-list negatives:", not_list[:10], "count:", len(not_list))
print("Non-string negatives:", non_str[:10], "count:", len(non_str))
print("Zero-length negatives:", length_zero[:10], "count:", len(length_zero))
```
